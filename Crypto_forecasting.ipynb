```python
df = pd.read_csv(path + 'BTC-USD.csv')

df.dropna(inplace= True)

df.head()
```

```python
#variable for predicting n days into the future
future_days = 1

#Add new column (target or dependent variable) shifted 'n' units up
df['Future_Price'] = df[['Close']].shift(-future_days)

#drop unneeded cols (not on this list)
df = df[['Date', 'Close', 'Future_Price']]

df.head()
```

```python
#Copy the dataset
df_copy = df.copy()

#Create independent dataset
X = np.array(df[df.columns])

#Remove the last 'n' rows from the dataset where 'n' is futre days
X = X[:len(df) - future_days]

#Create the dependent dataset (y) and convert dataset to numpy arr
y = np.array(df['Future_Price'])

#Get all the y values except for the future_days
y = y[:-future_days]

#Split the data into a training data and testing set
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.15, random_state = 0, shuffle = False)
```

```python
#Get train data and transform it to df
train_data = pd.DataFrame(X_train, columns = df.columns)

#Show the head
train_data.head()
```

```python
#Get the test data and transform it to df
test_data = pd.DataFrame(X_test, columns = df.columns)

#Show the head
test_data.head()
```

```python
#Initialize the setup
regression_setup = setup(data = train_data, target = 'Future_Price', session_id = 17, use_gpu = True)
```

```python
#Train on all of the models and sort it by the R-squared (a.k.a. r2),
#store the model with the best r2

best_model = compare_models(sort='r2')
```

```python
#Create the model and show its prediction metrics on the training data set
model = create_model(best_model)
```

```python
#Evaluate the model
evaluate_model(model)
```

```python
#Get the predictions
unseen_predictions = predict_model(model, data = test_data)

#Show the predictions
unseen_predictions
```
